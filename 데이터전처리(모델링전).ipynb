# test 고유 경기 번호 부여

```python
test = test.reset_index(drop=True)
test.head()
```

```python
test.shape
```

```python
unique_df = test[['rcDate', 'rcNo', 'meet2']].drop_duplicates().reset_index(drop=True)
unique_df.head()
```

```python
unique_df.shape
```

```python
test['unique_rcNo'] = np.nan
```

```python
def unique_rcNo(row):
  d = row['rcDate']
  n = row['rcNo']
  m = row['meet2']

  new = unique_df[(unique_df['rcDate']==d)&(unique_df['rcNo']==n)&(unique_df['meet2']==m)].index[0]

  return new
```

```python
test['unique_rcNo'] = test.apply(unique_rcNo, axis=1)
```

```python
test['unique_rcNo'].unique()
```

```python
test[test['unique_rcNo'].isna()]
```

```python
test = test.drop(['meet2'], axis=1)
```
